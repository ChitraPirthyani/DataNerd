```python


import scipy

import numpy as np
from scipy.stats import entropy
from sklearn.preprocessing import StandardScaler
from math import log, e
import pandas as pd


#Problem 2

Entropy_P= entropy([0.10,0.40,0.25,0.25], base=2)
print ("Entropy of P =", Entropy_P)

Entropy_Q = entropy([0.60,0.15,0.05,0.20], base=2)
print ("Entropy of Q =", Entropy_Q)

KL_Divergence_PQ = entropy(pk= ([0.10,0.40,0.25,0.25]), qk= ([0.60,0.15,0.05,0.20]), base=2)
print( "KL Divergence (P||Q) = ", KL_Divergence_PQ)

KL_Divergence_QP = entropy(pk=([0.60,0.15,0.05,0.20]) , qk= ([0.10,0.40,0.25,0.25]), base=2)
print( "KL Divergence (Q||P) = ", KL_Divergence_QP)

#Cross Entropy = Entropy + KL Divergence

CrossEntropy_PQ = Entropy_P + KL_Divergence_PQ
print ("Cross Entropy (P,Q)= ", CrossEntropy_PQ)

CrossEntropy_QP = Entropy_Q + KL_Divergence_QP
print ("Cross Entropy (Q,P)= ", CrossEntropy_QP)

```

```python
# Problem 3
# Writing PCA code from scratch

from numpy import array
from numpy import mean, cov
from numpy.linalg import eig

# Creating a matrix

Matrix = array([[2,5], [5,12], [9,18], [16,14], [3,6]])

print(Matrix)


# Column Means

ColMean = mean(Matrix.T, axis=1)
print ("Column Means:", ColMean)

# Centering Columns

A = Matrix - ColMean
print(A)

# Covariance of centered columns

CovA = cov(A.T)
print("Covariance of Centred cols:", )

# Eigen Values of Cov Matrix

values, vectors = eig(CovA)
print("vectors:", vectors)
print("values:", values)

#Final Data
Final = vectors.T.dot(A.T)
print("Final Principal Component:", Final.T)
```

```python
# Problem 4

import pandas as pd
url = "http://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data"

df = pd.read_csv(url,sep=",", header=None, index_col=None)
print(df)

# dropping the ID variable as it is irrelavant
df1= df.drop(df.index[0],axis=1)
df1.shape

# As PCA is only performed on numeric and not class variables, create a dataframe with numeric columns
numeric_data = df1[df1.columns[1:31]]
numeric_data.shape

target_data = df1[df1.columns[0]]

x = numeric_data.values
x = StandardScaler().fit_transform(x)

# Checking if Standard deviation is 1
np.std(x)
print("Standard Deviation:",np.std(x))

sampledata = x

# matrix multiplication

cov_matrix = np.matmul(sampledata.T, sampledata)

cov_matrix.shape

# Obtaining eigenvalues

from scipy.linalg import eigh

val, vect = eigh(cov_matrix, eigvals=(10,14))

vect = vect.T
vect.shape
print("PCA Shape after eigen value multiplication:",vect.shape)

# forming two principal eigen by vector multiplication

new_eig = np.matmul(vect, sampledata.T)
vect.shape

# Projecting PCA
new_eig = np.vstack(new_eig.T)
new_eig

dataframe= pd.DataFrame(data=new_eig, columns=("1st PC", "2nd PC", "3rd PC", "4th PC", "5th PC"))
print(dataframe)

# Resulting Data frame is PCA with 5 components



#### Logistic regression ####

## Test and Train split

train = df1.loc[0:400]
print(train)
train.shape

test = df1.loc[401:]
test.shape


# spliting target
x_train = train[train.columns[1:31]]
y_train = train[1]

x_test = test[test.columns[1:31]]
y_test = test[1]

from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# Creating the model
logistic_regression = LogisticRegression()

logistic_regression.fit(x_train, y_train)

# predicting of test
y_pred = logistic_regression.predict(x_test)


#calculating accuracy
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy_percentage = 100 * accuracy
print("Accuracy Percentage of logistic model:",accuracy_percentage)




#### Decision tree #####

from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

classifier = DecisionTreeClassifier()
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)

## confusion matrix

from sklearn.metrics import classification_report, confusion_matrix
print("Confusion Matrix:\n",confusion_matrix(y_test, y_pred))
print("Model Diagnostics:",classification_report(y_test, y_pred))



## Calculating ROC curves for both models

import matplotlib.pyplot as plt  
import seaborn as sns  
from sklearn.datasets import make_classification    
from sklearn.model_selection import train_test_split  
from sklearn.metrics import roc_curve

## Roc curve for Logistic regression model
from sklearn.metrics import plot_roc_curve
plot_roc_curve(logistic_regression, x_test, y_test, alpha=0.8, label= "ROC")
plt.title(label="Logistic Reg-ROC Curve")
plt.show()


## Roc curve for Decision tree model
plot_roc_curve(classifier, x_test, y_test, alpha=0.8, label= "ROC")
plt.title(label="Decision Tree-ROC Curve")
plt.show()

### Precision Recall curves ##
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import plot_precision_recall_curve

## Precision recall curve for Decision Tree
plot_precision_recall_curve(classifier, x_test, y_test, label="Precision Recall Curve")
plt.title(label="Decision Tree-Precision Recall Curve")
plt.show()

## Precision Recall curve for Logistic Regression Model
plot_precision_recall_curve(logistic_regression, x_test, y_test, label="Precision Recall Curve")
plt.title(label="Logistic Reg-Precision Recall Curve")
plt.show()

## MODEL METRICS ANALYSIS

print("MODEL ANALYSIS \n According to the various techniques applied, Logistic Regression is the best technique to classify this tye of data. \n The accuracy of the Logistic regression model: 93.4% was the highest of all three models. Decision Tree model gave the accuracy of 90%. \n While the ROC curves of both Logistic and Decision tree were both pretty favourable, the ROC of logistic regression performed slightly better. \n On the other hand, the Precision recall curve of the logistic regression performed significantly better than the decision tree's precision recall curve. The AUC values of both were 0.9 and 0.58 respectively. \n METRIC ANALYSIS \n As ROC curves are better suited when the observations are neatly balanced between each class, and Precision recall curves are more suited for imbalanced divisions, AUC is used as a model metric. \n As our data is clearly imbalanced between the two class, precision recall curves are a better suited metric for our data.\n When comparing Accuraacy scores, and Confusion Matrix, the latter is a more reliable indicator, as in addition to the accuracy percetage, it also shows other metrics like precision score and recall score etc.")


```
